### Capítulo 03. O Processo de Aprendizagem
#### Passos do Gradiente Descendente

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a função de custo
def cost_function(x):
    return (x + 5)**2

# Definir o gradiente da função de custo
def cost_function_gradient(x):
    return 2 * (x + 5)
```

```python
# Inicializar os parâmetros
x = 3.0
learning_rate = 0.01
max_iterations = 10000
tolerance = 1e-8

# Variáveis para armazenar o histórico dos valores
x_values = [x]
cost_values = [cost_function(x)]

# Inicializar variáveis para o loop
prev_x = x
iteration = 0
```

```python
# Aplicar gradiente descendente
while True:
    gradient = cost_function_gradient(x)
    new_x = x - learning_rate * gradient
    if abs(new_x - x) < tolerance:
        break
    x = new_x
    x_values.append(x)
    cost_values.append(cost_function(x))
    iteration += 1
    if iteration >= max_iterations:
        break
```

```python
# Resultados finais
print(f"Valor final de x: {x}")
print(f"Número de iterações: {iteration}")
```

```python
# Visualização do processo

# Gerar valores para a função de custo
x_range = np.linspace(-10, 5, 100)
y_range = cost_function(x_range)

# Plotar gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Função de custo
ax.plot(x_range, y_range, label='Função de Custo: $y=(x+5)^2$', color='blue')

# Caminho do gradiente descendente
ax.plot(x_values, cost_values, 'k-o', label='Caminho do Gradiente Descendente')

# Ponto inicial (vermelho) e ponto final (azul)
ax.plot(x_values[0], cost_values[0], 'ro', markersize=10, label='Ponto Inicial')
ax.plot(x_values[-1], cost_values[-1], 'bo', markersize=10, label='Ponto Final')

# Configurações do gráfico
ax.set_title('Visualização do Gradiente Descendente')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)

# Ajustar layout
fig.tight_layout()

# Salvar a figura em PNG
#fig.savefig('images/gradient_descent_simple_function.png', dpi=300)
plt.show()
```
