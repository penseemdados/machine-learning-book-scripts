# Descrição do Código do Gradiente Descendente
O código apresentado ilustra a aplicação do gradiente descendente para encontrar os coeficientes que minimizam uma função de custo em um contexto de *machine learning*. Abaixo, segue uma breve descrição de cada parte do código:

1. **Importação das Bibliotecas Necessárias**:

```python
import numpy as np
import matplotlib.pyplot as plt
```

2. **Definição da Função de Custo**:

```python
def cost_function(theta):
    return theta[0]**2 + theta[1]**2
```

A função de custo utilizada é $J(\theta) = \theta_0^2 + \theta_1^2$, uma função quadrática simples.

3. **Definição do Gradiente da Função de Custo**:

```python
def cost_function_gradient(theta):
    return np.array([2*theta[0], 2*theta[1]])
```

O gradiente da função de custo é calculado como $\nabla J(\theta) = [2 \theta_0, 2 \theta_1]$.

O símbolo $\nabla$ é conhecido como "nabla" ou "del" e é usado em matemática para denotar operadores diferenciais. No contexto do gradiente, $\nabla J(\theta)$ representa o gradiente da função de custo *J* em relação aos parâmetros $\theta$. O gradiente é um vetor que contém as derivadas parciais da função em relação a cada um dos seus parâmetros. Ele indica a direção da taxa de variação mais rápida da função, sendo fundamental para métodos de otimização como o gradiente descendente, onde é usado para ajustar os parâmetros a fim de minimizar a função de custo.

4. **Inicialização dos Parâmetros**:

```python
theta = np.array([4.0, 3.0])
learning_rate = 0.1
num_iterations = 50
```

Os parâmetros iniciais são definidos com um valor inicial de $\theta = [4.0,3.0]$, uma taxa de aprendizado ($\textit{learning rate}$) de 0.1 e o número de iterações em 50.

5. ***Arrays* para Armazenar os Valores das Iterações**:

```python
theta_values = [theta.copy()]
cost_values = [cost_function(theta)]
```

6. **Aplicação do Gradiente Descendente**:

```python
for _ in range(num_iterations):
    gradient = cost_function_gradient(theta)
    theta = theta - learning_rate * gradient
    theta_values.append(theta.copy())
    cost_values.append(cost_function(theta))

# Converter listas para arrays numpy para facilitar a plotagem
theta_values = np.array(theta_values)
cost_values = np.array(cost_values)
```

Um loop é utilizado para atualizar os valores de $\theta$ com base no gradiente da função de custo, ajustando os coeficientes a cada iteração para minimizar a função de custo.

7. **Geração do *Grid* de Valores para a Função de Custo**:

```python
theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)
theta0, theta1 = np.meshgrid(theta0_vals, theta1_vals)
J_vals = cost_function([theta0, theta1])
```

8. **Plotagem do Gráfico**:

```python
# Plotar gráfico
fig, ax = plt.subplots(figsize=(8, 6))

# Contorno da função de custo
CS = ax.contour(theta0, theta1, J_vals, levels=np.logspace(-1, 3, 20), cmap='viridis')
ax.clabel(CS, inline=1, fontsize=10)

# Caminho do gradiente descendente
ax.plot(theta_values[:, 0], theta_values[:, 1], 'k-', marker='o', label='Caminho do Gradiente Descendente')

# Ponto inicial (vermelho) e ponto final (azul)
ax.plot(theta_values[0, 0], theta_values[0, 1], 'ro', markersize=10, label='Ponto Inicial')
ax.plot(theta_values[-1, 0], theta_values[-1, 1], 'bo', markersize=10, label='Ponto Final')

# Configurações do gráfico
ax.set_title('Visualização do Gradiente Descendente')
ax.set_xlabel('Coeficiente 1')
ax.set_ylabel('Coeficiente 2')
ax.legend()
ax.grid(True)

# Ajustar layout para Kindle
fig.tight_layout()

# Salvar a figura em PNG
#fig.savefig('images/gradient_descent_visualization_corrected.png', dpi=300)
plt.show()
```

```python
# Plotar gráfico 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Superfície da função de custo
ax.plot_surface(theta0, theta1, J_vals, cmap='viridis', alpha=0.6)

# Caminho do gradiente descendente
ax.plot(theta_values[:, 0], theta_values[:, 1], cost_values, 'r-o', label='Caminho do Gradiente Descendente')

# Configurações do gráfico
ax.set_title('Visualização 3D do Gradiente Descendente')
ax.set_xlabel('Coeficiente 1')
ax.set_ylabel('Coeficiente 2')
ax.set_zlabel('Função de Custo')
ax.legend()

# Ajustar ângulo de visão
ax.view_init(elev=35, azim=240)

# Ajustar layout
fig.tight_layout()

# Salvar a figura em PNG
#fig.savefig('images/gradient_descent_visualization_3D.png', dpi=300)
plt.show()
```

O gráfico mostra o contorno da função de custo e o caminho do gradiente descendente desde o ponto inicial até o ponto final, ilustrando o processo de minimização da função de custo.

Essa descrição fornece uma visão geral das etapas e lógica implementadas no código, facilitando a compreensão do gradiente descendente e sua aplicação prática.
